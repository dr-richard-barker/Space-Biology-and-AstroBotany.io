```python
import requests
import json
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
# Get all files for OSD-37
response = requests.get('https://osdr.nasa.gov/genelab/data/glds/files/37').json()
```

```python
response
```

```python
# Get the remote URL for all files (need to put https://osdr.nasa.gov before each URL to access)
for item in response['studies']['OSD-37']['study_files']:
    print(item['remote_url'])
```

```python
# Use pandas to read in the Normalized Counts CSV file
pd.read_csv('https://osdr.nasa.gov/geode-py/ws/studies/OSD-37/download?source=datamanager&file=GLDS-37_rna_seq_Normalized_Counts.csv', nrows=10)
```

```python
# Use pandas to read in the Normalized Counts CSV file
pd.read_csv('https://osdr.nasa.gov/geode-py/ws/studies/OSD-38/download?source=datamanager&file=GLDS-38_rna_seq_Normalized_Counts.csv', nrows=10)
```

```python
from IPython.display import HTML

# Define the remote URL
remote_url = '/geode-py/ws/studies/OSD-38/download?source=datamanager&file=GLDS-38_rna_seq_Atha_WT-Col-0_sl_FLT_Rep1_G1S1_R1_raw_fastqc.html'

# Display the remote URL using HTML()
display(HTML(remote_url))
```

```python
# Load the HTML file
with open("/geode-py/ws/studies/OSD-38/download?source=datamanager&file=GLDS-38_rna_seq_Atha_WT-Col-0_sl_FLT_Rep1_G1S1_R1_raw_fastqc.html", "r") as f:
    html_content = f.read()

# Display the HTML content in the notebook
from IPython.display import HTML

display(HTML(html_content))
```
